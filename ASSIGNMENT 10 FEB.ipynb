```python
Q1. Which function is used to open a file? What are the different modes of opening a file? Explain each mode
of file opening.

ANS=The open() function is used to open a file in Python. It takes two parameters - the first parameter is the name of the file, and the second parameter is the mode in which to open the file.

r - read mode: This mode is used when the file needs to be read. The file is opened in read-only mode, and if the file does not exist, it raises a FileNotFoundError error.

w - write mode: This mode is used when the file needs to be written. If the file already exists, it will overwrite the existing file with the new data. If the file does not exist, it creates a new file.

a - append mode: This mode is used when data needs to be added to the end of the file. If the file does not exist, it creates a new file.

x - exclusive creation mode: This mode is used when a new file needs to be created, but if the file already exists, it raises a FileExistsError error.

b - binary mode: This mode is used when the file is a binary file. This mode is usually used in combination with the other modes, like rb, wb, and ab.

t - text mode: This mode is used when the file is a text file. This mode is usually used in combination with the other modes, like rt, wt, and at.

+ - read and write mode: This mode is used when the file needs to be read and written to. This mode is usually u

file = open("example.txt", "r")

file = open("example.txt", "w")

Q2. Why close() function is used? Why is it important to close a file?

ANS=The close() function is used to close a file that was previously opened using the open() function in Python. It is important to close a file after it has been opened and used because it frees up system resources and ensures that any data in the buffer is written to the file.
To avoid these problems, it is important to close a file after it has been opened and used. The close() function can be called on the file object to close the file and free up system resources. For example, to close a file that was opened in read mode, you can use the following code:
file = open("example.txt", "r")
# do something with the file
file.close()

Q3. Write a python program to create a text file. Write ‘I want to become a Data Scientist’ in that file. Then
close the file. Open this file and read the content of the file.
ANS=Here's a Python program that creates a text file, writes the string "I want to become a Data Scientist" to the file, closes the file, opens it again, and reads the contents of the file:
# Open the file in write mode
file = open("data_scientist.txt", "w")

# Write the string to the file
file.write("I want to become a Data Scientist")

# Close the file
file.close()

# Open the file in read mode
file = open("data_scientist.txt", "r")

# Read the contents of the file
contents = file.read()

# Close the file
file.close()

# Print the contents of the file
print(contents)
Q4. Explain the following with python code: read(), readline() and readlines().
ANS=read(): The read() method is used to read the entire contents of a file as a single string. This method takes an optional parameter that specifies the number of characters to read. If no parameter is given, the entire file is read. Here's an example:
# Open the file in read mode
file = open("example.txt", "r")

# Read the entire contents of the file
contents = file.read()

# Close the file
file.close()

# Print the contents of the file
print(contents)
readline(): The readline() method is used to read a single line from a file. This method reads up to and including the first newline character ("\n") and returns a string. If the end of the file is reached, an empty string is returned. Here's an example:# Open the file in read mode
file = open("example.txt", "r")

# Read the first line of the file
line = file.readline()

# Close the file
file.close()

# Print the line
print(line)
readlines(): The readlines() method is used to read all the lines of a file into a list. Each line is a separate element in the list. Here's an example:# Open the file in read mode
file = open("example.txt", "r")

# Read all the lines of the file into a list
lines = file.readlines()

# Close the file
file.close()

# Print each line
for line in lines:
    print(line)

Q5. Explain why with statement is used with open(). What is the advantage of using with statement and
open() together?

The with statement in Python is used to create a context in which a file or other resource is automatically managed. When used with the open() function, it ensures that the file is closed properly after it has been used.
# Open the file in read mode using with statement
with open('example.txt', 'r') as file:
    # Read the entire contents of the file
    contents = file.read()

# The file is automatically closed when the block is exited
# Print the contents of the file
print(contents)
In this example, the with statement is used to open the file "example.txt" in read mode. The with statement automatically closes the file when the block of code under the with statement is exited. The contents of the file are read using the read() method and stored in the contents variable. Finally, the contents

Q6. Explain the write() and writelines() functions. Give a suitable example.








```
